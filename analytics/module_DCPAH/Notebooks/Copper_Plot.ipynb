```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import pylab

pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
cell_area = []
f = open('Copper.txt', 'r')
for line in f:
    if '.tif' in line:
        words = line.split()
        if (float(words[1]) > 1):
            cell_area.append(float(words[1]))
            
plt.hist(cell_area, bins=np.arange(min(cell_area), max(cell_area)+5, 100))
```
