```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pylab
import os

from skimage.feature import peak_local_max
from skimage.morphology import watershed
from scipy import ndimage

pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
gray_img1 = cv2.imread("../Images/SNAP-084625-0008.jpg",0)
gray_img2 = cv2.imread('../Images/SNAP-152832-0032.jpg',0)
gray_img3 = cv2.imread('../Images/SNAP-153730-0038.jpg',0)

img1 = cv2.imread("../Images/SNAP-084625-0008.jpg")
img2 = cv2.imread('../Images/SNAP-152832-0032.jpg')
img3 = cv2.imread('../Images/SNAP-153730-0038.jpg')

image1 = cv2.GaussianBlur(img1, (0,0), 3)
image2 = cv2.GaussianBlur(img2, (0,0), 3)
image3 = cv2.GaussianBlur(img3, (0,0), 3)
img1 = cv2.addWeighted(img1, 1.5, image1, -0.5, 0)
img2 = cv2.addWeighted(img2, 1.5, image2, -0.5, 0)
img3 = cv2.addWeighted(img3, 1.5, image3, -0.5, 0)

plt.subplot(321), plt.imshow(img1), plt.xticks([]), plt.yticks([])
plt.subplot(322), plt.imshow(gray_img1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(323), plt.imshow(img2), plt.xticks([]), plt.yticks([])
plt.subplot(324), plt.imshow(gray_img2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(325), plt.imshow(img3), plt.xticks([]), plt.yticks([])
plt.subplot(326), plt.imshow(gray_img3, 'gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Adaptive threshold did not help 
thresh = cv2.adaptiveThreshold(gray_img1, 70, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 0)
#plt.imshow(thresh, 'gray')
#plt.show()
```

```python
# grayscale -> manual thresh
_, thresh1 = cv2.threshold(gray_img1, 70, 255, cv2.THRESH_BINARY)
_, thresh2 = cv2.threshold(gray_img2, 70, 255, cv2.THRESH_BINARY)
_, thresh3 = cv2.threshold(gray_img3, 70, 255, cv2.THRESH_BINARY)
plt.subplot(321), plt.imshow(gray_img1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(322), plt.imshow(255-thresh1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(323), plt.imshow(gray_img2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(324), plt.imshow(255-thresh2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(325), plt.imshow(gray_img3, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(326), plt.imshow(255-thresh3, 'gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
yuv_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2YUV)
yuv_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2YUV)
yuv_img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2YUV)

thresh1 = cv2.inRange(yuv_img1, (0,0,0), (75,255,255))
thresh2 = cv2.inRange(yuv_img2, (0,0,0), (75,255,255))
thresh3 = cv2.inRange(yuv_img3, (0,0,0), (75,255,255))
plt.subplot(321), plt.imshow(yuv_img1), plt.xticks([]), plt.yticks([])
plt.subplot(322), plt.imshow(thresh1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(323), plt.imshow(yuv_img2), plt.xticks([]), plt.yticks([])
plt.subplot(324), plt.imshow(thresh2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(325), plt.imshow(yuv_img3), plt.xticks([]), plt.yticks([])
plt.subplot(326), plt.imshow(thresh3, 'gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
mask1 = np.zeros(thresh1.shape, np.uint8)
mask2 = np.zeros(thresh2.shape, np.uint8)
mask3 = np.zeros(thresh3.shape, np.uint8)

c1, _ = cv2.findContours(thresh1, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
c2, _ = cv2.findContours(thresh2, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
c3, _ = cv2.findContours(thresh3, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
cells = []
for cnt in c1:
    area = cv2.contourArea(cnt)
    print(area)
    if area > 100 and area < 5000:
        cv2.drawContours(mask1, [cnt], 0, 255, -1)
        cells.append(cnt)

print("LEN OF CELLS: {}".format(len(cells)))
plt.imshow(mask1, 'gray'), plt.show()
cells = []
for cnt in c2:
    area = cv2.contourArea(cnt)
    print(area)
    if area > 100 and area < 5000:
        cv2.drawContours(mask2, [cnt], 0, 255, -1)
        cells.append(cnt)

print("LEN OF CELLS: {}".format(len(cells)))
plt.imshow(mask2, 'gray'), plt.show()

cells = []
for cnt in c3:
    area = cv2.contourArea(cnt)
    print(area)
    if area > 100 and area < 5000:
        cv2.drawContours(mask3, [cnt], 0, 255, -1)
        cells.append(cnt)

print("LEN OF CELLS: {}".format(len(cells)))
plt.imshow(mask3, 'gray'), plt.show()

plt.show()
```

```python
plt.subplot(131), plt.imshow(mask1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(mask2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(mask3, 'gray'), plt.xticks([]), plt.yticks([])
plt.show()

m1 = np.zeros(mask1.shape, np.uint8)
m2 = np.zeros(mask2.shape, np.uint8)
m3 = np.zeros(mask3.shape, np.uint8)

new_img1 = cv2.bitwise_and(img1, img1, mask=mask1)
new_img2 = cv2.bitwise_and(img2, img3, mask=mask2)
new_img3 = cv2.bitwise_and(img2, img3, mask=mask3)

plt.subplot(131), plt.imshow(new_img1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(new_img2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(new_img3, 'gray'), plt.xticks([]), plt.yticks([])
plt.show()

cv2.imwrite('mask_color_area_thresh_1.png', new_img1)
cv2.imwrite('mask_color_area_thresh_2.png', new_img2)
cv2.imwrite('mask_color_area_thresh_3.png', new_img3)
```

```python
dist = cv2.distanceTransform(mask1, cv2.cv.CV_DIST_L2, 5)
localMax = peak_local_max(dist, indices=False, min_distance=20, labels=mask1)
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-dist, markers, mask=mask1)
plt.imshow(labels), plt.show()
for label in np.unique(labels):
    if label == 0:
        continue
    labelled = np.zeros(dist.shape, dtype="uint8")
    labelled[labels == label] = 255
    cnts = cv2.findContours(labelled, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
    for c in cnts:
        cv2.drawContours(m1, [c],0,255,-1)
        
dist = cv2.distanceTransform(mask2, cv2.cv.CV_DIST_L2, 5)
localMax = peak_local_max(dist, indices=False, min_distance=20, labels=mask2)
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-dist, markers, mask=mask2)
plt.imshow(labels), plt.show()
for label in np.unique(labels):
    if label == 0:
        continue
    labelled = np.zeros(dist.shape, dtype="uint8")
    labelled[labels == label] = 255
    cnts = cv2.findContours(labelled, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
    for c in cnts:
        cv2.drawContours(m2, [c],0,255,-1)
        
dist = cv2.distanceTransform(mask3, cv2.cv.CV_DIST_L2, 5)
localMax = peak_local_max(dist, indices=False, min_distance=20, labels=mask3)
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-dist, markers, mask=mask3)
plt.imshow(labels), plt.show()
for label in np.unique(labels):
    if label == 0:
        continue
    labelled = np.zeros(dist.shape, dtype="uint8")
    labelled[labels == label] = 255
    cnts = cv2.findContours(labelled, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
    for c in cnts:
        cv2.drawContours(m3, [c],0,255,-1)
        
plt.subplot(131), plt.imshow(m1, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(m2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(m3, 'gray'), plt.xticks([]), plt.yticks([])
plt.show()
```
