```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pylab
import os

pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
# -------------- TEST 1 ---------------------------------------------------
file_path = os.listdir("../Images/")
copper_test = open('Copper.txt', 'w')
copper_test.write("Image\tTotal Cell Area\tTissue Area\tNormalized(Cell/Tissue)\n")
for infile in file_path:
    if "SP1" in infile:
        img = cv2.imread('../Images/'+infile)
        orig = img.copy()
        tissue_area = 1
        mask = cv2.inRange(img, (0, 0, 165), (155,170,200))
        c, _ = cv2.findContours(mask, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        cells = []
        cell_area = 0
        for cnt in c:
            if cv2.contourArea(cnt) > 5:
                cells.append(cnt)
                cell_area += cv2.contourArea(cnt)
                cv2.drawContours(img, [cnt], -1, (255,0,0) ,3)
            
            
        cv2.imwrite('Copper_Test1_Orig/Copper_'+infile, img)
        cv2.imwrite('Copper_Test1_Mask/Copper_'+infile, mask)
        
        copper_test.write(infile+'\t'+str(cell_area)+'\t'+str(tissue_area)+'\t'+str(cell_area/tissue_area)+'\n')
        #plt.subplot(121), plt.imshow(orig), plt.title('Original'), plt.xticks([]), plt.yticks([])
        #plt.subplot(122), plt.imshow(img), plt.title(infile), plt.xticks([]), plt.yticks([])
        #plt.show()
        
copper_test.close()
```

```python
# ------ COMBINE TEST1 + TEST2 TO ELIMINATE FALSE POSITIVE CELLS ------------------------

file_path = os.listdir("../Images/")
copper_test = open('Copper3.txt', 'w')
copper_test.write("Image\tTotal Cell Area\tTissue Area\tNormalized(Cell/Tissue)\n")
for infile in file_path:
    if "SP1" in infile:
        img = cv2.imread('../Images/'+infile)
        orig = img.copy()
        tissue_area = 1
        mask = cv2.inRange(img, (0, 0, 165), (155,160,200))
        mask2 = cv2.inRange(img, (0,0,165), (155, 120, 200))
        c, _ = cv2.findContours(mask, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        c2, _ = cv2.findContours(mask2, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        cells = []
        cell_area = 0
        if len(c2) > 10:
            for cnt in c:
                if cv2.contourArea(cnt) > 5:
                    cells.append(cnt)
                    cell_area += cv2.contourArea(cnt)
                    cv2.drawContours(img, [cnt], -1, (255,0,0) ,3)
        else:
            for cnt in c2:
                cells.append(cnt)
                cell_area += cv2.contourArea(cnt)
                cv2.drawContours(img, [cnt], -1, (0,0,255), 3)    

        cv2.imwrite('Copper_Test3_Orig/Copper_'+infile, img)
        cv2.imwrite('Copper_Test3_Mask/Copper_'+infile, mask)
        
        copper_test.write(infile+'\t'+str(cell_area)+'\t'+str(tissue_area)+'\t'+str(cell_area/tissue_area)+'\n')
        #plt.subplot(121), plt.imshow(orig), plt.title('Original'), plt.xticks([]), plt.yticks([])
        #plt.subplot(122), plt.imshow(img), plt.title(infile), plt.xticks([]), plt.yticks([])
        #plt.show()
        
copper_test.close()
```

```python
# --------------------------- TESTING TISSUE NORMALIZATION ----------------------------------
from itertools import groupby

max_key = []
file_path = os.listdir("../Images/")
for infile in file_path:
    if "Ki67" in infile:
        img = cv2.imread('../Images/'+infile)
        b,g,r = cv2.split(img)
        rgb_img = cv2.merge([r,g,b])
        rg = cv2.merge([r,g,b])
        rg = rg.astype(np.float32)

        np.seterr(all = 'ignore')
        mask = np.empty_like(rg)
        arr_max = rg.max(-1)
        delta = rg.ptp(-1)
        s = delta/arr_max
        s[delta==0]=0
        idx = (rg[:,:,0] == arr_max)
        mask[idx,0] = np.true_divide((rg[idx,1] - rg[idx,2]), delta[idx])
        idx = (rg[:,:,1] == arr_max)
        mask[idx,0] = 2. + np.true_divide((rg[idx,2] - rg[idx,0]), delta[idx])
        idx = (rg[:,:,2] == arr_max)
        mask[idx,0] = 4. + np.true_divide((rg[idx,0] - rg[idx,1]), delta[idx])
        mask[:,:,0] = ((mask[:,:,0]/6.0)%1.0)*255.0
        mask[:,:,0] = np.nan_to_num(mask[:,:,0])
        mask[:,:,1] = s * 255.0
        mask[:,:,2] = arr_max
        mask = mask.astype(np.uint8)

        freq = {key: len(list(group)) for key, group in groupby(np.sort(mask[:,:,0][0]))}
        max_key.append(max(freq, key=lambda k: freq[k]))

print(np.mean(max_key))
```
