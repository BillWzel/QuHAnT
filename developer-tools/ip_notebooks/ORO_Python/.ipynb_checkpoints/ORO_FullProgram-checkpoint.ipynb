```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def QuHAnT(file_path, sthresh, bthresh, min_feat, id_file, ext, type_file, savei, in_file):
# Function assesses which option was selected and sends to correct functions.
# Writes files based on given results

    i_files = []
    o_files = []

    file_name = os.listdir(file_path)
    
    # checks if option 1 or 2 was selected
    if type_file == 1 or type_file == 2:
        (i_files, o_files) = FastImageLabel(file_path, ext, sthresh, type_file, id_file)
        
    elif type_file == 3:
        # Gen_FileList generates array
        i_files = Gen_FileList(file_path, ext, id_file)

    elif type_file == 4:
        dir_list = []
        # Loops through all the images in selected folder
        for infile in file_name:
            # Checks if extension of item matches the inputted extension
            if (infile[-4:] == ext): 
                # Appends to array
                dir_list.append(infile)
        # Loops through newly formed array
        for i in range(len(dir_list)):
            # If number in given matrix, append to final array
            if i in in_file:
                i_files.append(dir_list[i])
    else:
        # assert error if input was not 1-4
        assert(1), "Not a valid input"
      
    # Create empty dictionaries
    s_dict = {}
    b_dict = {}
    f_summary = {}
    b_summary = {}
    
    # Start creating files
    fs = open('Feature_Summary.txt', 'w')
    fs.write("\nFeature Summary\n\n")
    f_ind =open('Individual_Features.txt', 'w')
    f_ind.write("\nIndividual Features\n\n")
    back_summ =open('Background_Summary.txt', 'w')
    back_summ.write("\nBackground Summary\n\n")
    b_ind =open('Background_Features.txt', 'w')
    b_ind.write("\nBackground Features\n\n")
    
    # For each image, send to be quantified and write results into files
    for ifile in i_files:
        if ifile != 'na':
            (s_dict, b_dict, f_summary, b_summary) = QuHAnTitation(ifile, sthresh, bthresh, min_feat, savei)
    
            # Write quanitified results into files
            for key in sorted(f_summary):
                summ = str(f_summary[key])
                if (key[1] == 6):
                    fs.write(summ + '\n')
                else:
                    fs.write(summ + '\t')
            
            for key in sorted(s_dict):
                summ1 = str(s_dict[key])
                if (key[1] == 8):
                    f_ind.write(summ1 + '\n')
                else:
                    f_ind.write(summ1 + '\t')
                    
            for key in sorted(b_summary):
                summ2 = str(b_summary[key])
                if (key[1] == 6):
                    back_summ.write(summ2 + '\n')
                else:
                    back_summ.write(summ2 + '\t')
            
            for key in sorted(b_dict):
                summ3 = str(b_dict[key])
                if (key[1] == 8):
                    b_ind.write(summ3 + '\n')
                else:
                    b_ind.write(summ3 + '\t')
            
        else:
            # Write headers into files 
            fs =open('Feature_Summary.txt', 'w')
            fs.write("\nFeature Summary\n\n")
            fs.write("na")
            
            f_ind =open('Individual_Features.txt', 'w')
            f_ind.write("\nIndividual Features\n\n")
            f_ind.write("na")
            
            back_summ =open('Background_Summary.txt', 'w')
            back_summ.write("\nBackground Summary\n\n")
            back_summ.write("na")
            
            b_ind =open('Background_Features.txt', 'w')
            b_ind.write("\nBackground Features\n\n")
            b_ind.write("na")
            
    # Close files
    fs.close()
    f_ind.close()
    back_summ.close()
    b_ind.close()
            
```

```python
def Show_Image(file_path, sthresh):
# Thresholds the image and displays the image and mask

    im=cv2.imread(file_path)
    b,g,r = cv2.split(im)
    rgb_img = cv2.merge([r,g,b])
    rg = cv2.merge([r,g,b])
    rg = rg.astype(np.float32)

    # Masking Image
    # Source code: https://github.com/scikit-image/scikit-image/blob/master/skimage/color/colorconv.py

    np.seterr(all = 'ignore')
    mask = np.empty_like(rg)
    arr_max = rg.max(-1)
    delta = rg.ptp(-1)
    s = delta/arr_max
    s[delta==0]=0
    idx = (rg[:,:,0] == arr_max)
    mask[idx,0] = np.true_divide((rg[idx,1]-rg[idx,2]), delta[idx])
    idx = (rg[:,:,1] == arr_max)
    mask[idx,0] = 2. + np.true_divide((rg[idx,2]-rg[idx,0]), delta[idx])
    idx = (rg[:,:,2] == arr_max)
    mask[idx,0] = 4. + np.true_divide((rg[idx,0]-rg[idx,1]), delta[idx])
    mask[:,:,0] = ((mask[:,:,0]/6.0)% 1.0)*255.0
    mask[:,:,0] = np.nan_to_num(mask[:,:,0])
    mask[:,:,1] = s * 255.0
    mask[:,:,2] = arr_max
    mask=mask.astype(np.uint8)
    
    # Thresholding Image
    
    R = sthresh
    if R[0][0] > R[0][1]:
        m1 = np.logical_or(R[0][0] <= mask[:,:,0], mask[:,:,0] <= R[0][1])
    else:
        m1 = np.logical_and(R[0][0] <= mask[:,:,0], mask[:,:,0] <= R[0][1])

    if R[1][0] > R[1][1]:
        m2 = np.logical_or(R[1][0] <= mask[:,:,1], mask[:,:,1] <= R[1][1])
    else:
        m2 = np.logical_and(R[1][0] <= mask[:,:,1], mask[:,:,1] <= R[1][1])

    if R[2][0] > R[2][1]:
        m3 = np.logical_or(R[2][0] <= mask[:,:,2], mask[:,:,2] <= R[2][1])
    else:
        m3 = np.logical_and(R[2][0] <= mask[:,:,2], mask[:,:,2] <= R[2][1])

    valid = np.logical_and(m1, m2)
    valid_range = np.logical_and(valid, m3)
    
    mask[valid_range] = 0
    mask[np.logical_not(valid_range)] = 255
        
    # Displays the original and masked image
    titles = ['Original', 'Masked']
    images = [rgb_img, mask]
    for i in xrange(2):
        plt.subplot(1,2,i+1), plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
def Gen_FileList(file_path, ext, id_file):
# Goes through the directory and if extension and id matches, gets appended to final array
    file_name = os.listdir(file_path)

    keep = []
    for infile in file_name:
        # if extension and id match the image name
        if (infile[-4:] == ext and id_file in infile):
            keep.append(infile)
            
    return keep
```

```python
def FastImageLabel(file_path, ext, sthresh, type_file, id_file):
# Displays the images and waits for user input on whether image should be kept or ignored

    i = 0
    flag = []

    file_name = os.listdir(file_path)

    dir_list = []
    for infile in file_name:
        # If extension and id match the image name, add to array
        if (infile[-4:] == ext and id_file in infile):
            dir_list.append(infile)

    # Create an array of len dir_list that is initialized to zeros
    for m in range(len(dir_list)):
        flag.append(0)
    
    ans = ''
    i = 0
    # While the user hasn't exited the program
    while (ans != 'q'):
        # Checks the option initially selected
        if (type_file == 1):
            # Displays the original/masked image to user
            im = cv2.imread(dir_list[i])
            Show_Image(dir_list[i], sthresh)

        elif (type_file == 2):
            # Shows only the image 
            im = cv2.imread(dir_list[i])
            b,g,r = cv2.split(im)
            rgb_img = cv2.merge([r,g,b])
            plt.imshow(rgb_img)
            plt.title(dir_list[i])
            plt.show()

        # Checks the user input
        ans = input("Quit('q'), Next ('n'), Previous ('p') or Keep ('k'): ")
        if ans == 'q':
            break
        elif ans == 'n':
            i += 1
        elif ans == 'p':                
            i -= 1
        elif ans == 'k':
            flag[i] = 1
            i+=1
        else:
            print("Invalid input")
            
        if i < 0:
            i = 0
        elif i == len(dir_list):
            break
                
    keep = []
    kick = []
    # Based on whether the flag was changed, puts into keep or kick array
    for k in range(len(flag)):
        if flag[k] == 1:
            keep.append(dir_list[k])
            kick.append('na')
        else:
            kick.append(dir_list[k])
            keep.append('na')
            
            
    return (keep, kick)
```

```python
def QuHAnTitation(file_path, sthresh, bthresh, min_feat, savei):
    # Quantitates the image through color conversion, thresholding and area analysis
    
    s_dict = {}
    s_calc = []
    s_clust1 = []
    s_clust2 = []
    b_dict = {}
    b_calc = []
    b_clust1 = []
    b_clust2 = []
    
    # Reads in image and converts from bgr to rgb
    im=cv2.imread(file_path)
    b,g,r = cv2.split(im)
    rgb_img = cv2.merge([r,g,b])
    rg = cv2.merge([r,g,b])
    rg = rg.astype(np.float32)

    # Masks the image
    # Source code: https://github.com/scikit-image/scikit-image/blob/master/skimage/color/colorconv.py#L159
    
    np.seterr(all = 'ignore')
    mask = np.empty_like(rg)
    arr_max = rg.max(-1)
    delta = rg.ptp(-1)
    s = delta/arr_max
    s[delta==0]=0
    idx = (rg[:,:,0] == arr_max)
    mask[idx,0] = np.true_divide((rg[idx,1]-rg[idx,2]), delta[idx])
    idx = (rg[:,:,1] == arr_max)
    mask[idx,0] = 2. + np.true_divide((rg[idx,2]-rg[idx,0]), delta[idx])
    idx = (rg[:,:,2] == arr_max)
    mask[idx,0] = 4. + np.true_divide((rg[idx,0]-rg[idx,1]), delta[idx])
    mask[:,:,0] = ((mask[:,:,0]/6.0)% 1.0)*255.0
    mask[:,:,0] = np.nan_to_num(mask[:,:,0])
    mask[:,:,1] = s * 255.0
    mask[:,:,2] = arr_max
    s_mask=mask.astype(np.uint8)

    # Thresholds the image based on foreground thresholds
    R = sthresh
    if R[0][0] > R[0][1]:
        m1 = np.logical_or(R[0][0] <= s_mask[:,:,0], s_mask[:,:,0] <= R[0][1])
    else:
        m1 = np.logical_and(R[0][0] <= s_mask[:,:,0], s_mask[:,:,0] <= R[0][1])

    if R[1][0] > R[1][1]:
        m2 = np.logical_or(R[1][0] <= s_mask[:,:,1], s_mask[:,:,1] <= R[1][1])
    else:
        m2 = np.logical_and(R[1][0] <= s_mask[:,:,1], s_mask[:,:,1] <= R[1][1])

    if R[2][0] > R[2][1]:
        m3 = np.logical_or(R[2][0] <= s_mask[:,:,2], s_mask[:,:,2] <= R[2][1])
    else:
        m3 = np.logical_and(R[2][0] <= s_mask[:,:,2], s_mask[:,:,2] <= R[2][1])

    valid = np.logical_and(m1, m2)
    valid_range = np.logical_and(valid, m3)

    s_mask[valid_range] = 255
    s_mask[np.logical_not(valid_range)] = 0
    
    s_grey = cv2.cvtColor(s_mask, cv2.COLOR_BGR2GRAY)
    s_result = cv2.bitwise_and(im,im,mask=s_grey)
        
    # Thresholds the image based on background thresholds
    b_mask=mask.astype(np.uint8)
    R = bthresh
    if R[0][0] > R[0][1]:
        m1 = np.logical_or(R[0][0] <= b_mask[:,:,0], b_mask[:,:,0] <= R[0][1])
    else:
        m1 = np.logical_and(R[0][0] <= b_mask[:,:,0], b_mask[:,:,0] <= R[0][1])

    if R[1][0] > R[1][1]:
        m2 = np.logical_or(R[1][0] <= b_mask[:,:,1], b_mask[:,:,1] <= R[1][1])
    else:
        m2 = np.logical_and(R[1][0] <= b_mask[:,:,1], b_mask[:,:,1] <= R[1][1])

    if R[2][0] > R[2][1]:
        m3 = np.logical_or(R[2][0] <= b_mask[:,:,2], b_mask[:,:,2] <= R[2][1])
    else:
        m3 = np.logical_and(R[2][0] <= b_mask[:,:,2], b_mask[:,:,2] <= R[2][1])

    valid = np.logical_and(m1, m2)
    valid_range = np.logical_and(valid, m3)

    b_mask[valid_range] = 255
    b_mask[np.logical_not(valid_range)] = 0
    b_grey=cv2.cvtColor(b_mask,cv2.COLOR_BGR2GRAY)
        
    s_contours, s_hierarchy = cv2.findContours(s_grey, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    i = 0
    # Analyze each contour to find specific information (bounding box, centroids, area)
    for cnt in s_contours:
        i+=1
        j = 1
        moment = cv2.moments(cnt)
        # If area is not zero
        if moment['m00'] != 0.0:
            cx = moment['m10']/moment['m00']
            cy = moment['m01']/moment['m00']
            centroid = (cx,cy)
        else:
            centroid = "Region has zero area"
            cx = 0
            cy = 0
        vertices = cv2.boundingRect(cnt)
        if cv2.contourArea(cnt) >= min_feat:
            s_dict[i,j] = file_path
            s_dict[i, j+1] = cv2.contourArea(cnt)
            s_dict[i, j+2] = cx
            s_dict[i, j+3] = cy
            s_dict[i, j+4] = vertices[0]
            s_dict[i, j+5] = vertices[1]
            s_dict[i, j+6] = vertices[2]
            s_dict[i, j+7] = vertices[3]
            s_calc.append(cv2.contourArea(cnt))
            s_clust1.append(cx)
            s_clust2.append(cy)
        
    f_summary = {}
    j = 1
    i = 1
    (mean, stddev) = cv2.meanStdDev(np.array(s_calc))
    if len(s_calc) != 0:
        f_summary[i, j] = file_path
        f_summary[i, j + 1] = len(s_contours)
        f_summary[i, j + 2] = sum(s_calc)
        f_summary[i, j + 3] = mean[0][0]
        f_summary[i, j + 4] = np.median(s_calc)
        f_summary[i, j + 5] = stddev[0][0]
    else:
        f_summary[i, j] = file_path
        f_summary[i, j + 1] = 0
        f_summary[i, j + 2] = 0
        f_summary[i, j + 3] = 0
        f_summary[i, j + 4] = 0
        f_summary[i, j + 5] = 0
        
    # Analyze the background contours
    
    b_contours, b_hierarchy = cv2.findContours(b_grey, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    i = 0
    for cnt in b_contours:
        i+=1
        j = 1
        moment = cv2.moments(cnt)
        if moment['m00'] != 0.0:
            cx = moment['m10']/moment['m00']
            cy = moment['m01']/moment['m00']
            centroid = (cx,cy)
        else:
            centroid = "Region has zero area"
            cx = 0 
            cy = 0
        vertices = cv2.boundingRect(cnt)
        if cv2.contourArea(cnt) >= min_feat:
            b_dict[i,j] = file_path
            b_dict[i, j+1] = cv2.contourArea(cnt)
            b_dict[i, j+2] = cx
            b_dict[i, j+3] = cy
            b_dict[i, j+4] = vertices[0]
            b_dict[i, j+5] = vertices[1]
            b_dict[i, j+6] = vertices[2]
            b_dict[i, j+7] = vertices[3]
            b_calc.append(cv2.contourArea(cnt))
            b_clust1.append(cx)
            b_clust2.append(cy)
        
    j = 1
    i = 1
    b_summary = {}
    (mean, stddev) = cv2.meanStdDev(np.array(b_calc))
    if len(b_calc) != 0:
        b_summary[i,j] = file_path
        b_summary[i, j + 1] = len(b_contours)
        b_summary[i, j + 2] = sum(b_calc)
        b_summary[i, j + 3] = mean[0][0]
        b_summary[i, j + 4] = np.median(b_calc)
        b_summary[i, j + 5] = stddev[0][0]
    else:
        b_summary[i, j] = file_path
        b_summary[i, j + 1] = 0
        b_summary[i, j + 2] = 0
        b_summary[i, j + 3] = 0
        b_summary[i, j + 4] = 0
        b_summary[i, j + 5] = 0
        
    # If option to save was selected, create mask
    if savei == 'y':
        name = str(file_path[:-4])
        cv2.imwrite( name + "_mask.png", s_result)
        
    return (s_dict, b_dict, f_summary, b_summary)
```
