```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Function tests fourier method

im = cv2.imread("../Images/GImage22.tif", 0)
dft = cv2.dft(np.float32(im), flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

magnitude_s = 20*np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))
plt.subplot(121), plt.imshow(im, cmap='gray')
plt.title("Input Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_s, cmap='gray')
plt.title("Magnitude Spectrum"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Function masks out everything but the middle of the fourier image

mask = np.zeros(im.shape, np.uint8)
mask[341:682,426:853] = magnitude_s[341:682,426:853]
plt.imshow(mask)
plt.show()
```

```python
#Function displays information about the images above

(mean, stddev) = cv2.meanStdDev(np.array(mask[341:682,426:853]))
(mean1, stddev1) = cv2.meanStdDev(np.array(magnitude_s))

print "Ratio: ", sum(sum(mask[341:682,426:853]))/sum(sum(magnitude_s))
print "Sum of whole image:",(sum(sum(magnitude_s)))
print "Sum of middle: ",(sum(sum(mask[341:682,426:853])))
print "Mean of image: ", mean1[0][0]
print "Mean of middle: ", mean[0][0]
print "Median of image: ", np.median(magnitude_s)
print "Median of middle: ", np.median(mask[341:682,426:853])
print "Std Dev of image: ", stddev1[0][0]
print "Std Dev of middle: ", stddev[0][0]
print "Min of image: ", np.min(np.array(magnitude_s))
print "Min of middle: ", np.min(np.array(mask[341:682,426:853]))
print "Max of image: ", np.max(np.array(magnitude_s))
print "Max of middle: ", np.max(np.array(mask[341:682,426:853]))
print "Difference of image: ", np.max(np.array(magnitude_s))-np.min(np.array(magnitude_s))
print "Difference of middle: ", np.max(np.array(mask[341:682,426:853]))-np.min(np.array(mask[341:682,426:853]))
#print "Laplacian of image: ", cv2.Laplacian(magnitude_s, cv2.CV_8U).var()
#print "Laplacian of middle: ", cv2.Laplacian(mask[341:682,426:853], cv2.CV_8U).var()
```

```python
#Function tests Fourier and histograms

img = magnitude_s
mask = np.zeros(img.shape, np.uint8)
mask[300:700,400:900] = 255
masked_img = cv2.bitwise_and(img, img, mask = mask)

hist_full = cv2.calcHist([img], [0], None, [256], [0,256])
hist_mask = cv2.calcHist([img], [0], mask, [256], [0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask, 'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```

```python
#www.astrobetter.com/blog/2010/03/03/fourier-transforms-of-images-in-python

import cv2
from scipy import fftpack
#import pyfits
import numpy as np
import pylab as py
#import radialProfile
 
image = cv2.imread('../Images/Image20.tif')
 
# Take the fourier transform of the image.
F1 = fftpack.fft2(image)
 
# Now shift the quadrants around so that low spatial frequencies are in
# the center of the 2D fourier transformed image.
F2 = fftpack.fftshift( F1 )
 
# Calculate a 2D power spectrum
psd2D = np.abs( F2 )**2
 
# Calculate the azimuthally averaged 1D power spectrum
psd1D = radialProfile.azimuthalAverage(psd2D)
 
# Now plot up both
py.figure(1)
py.clf()
py.imshow( np.log10( image ), cmap=py.cm.Greys)
 
py.figure(2)
py.clf()
py.imshow( np.log10( psf2D ))
 
py.figure(3)
py.clf()
py.semilogy( psf1D )
py.xlabel('Spatial Frequency')
py.ylabel('Power Spectrum')
 
py.show()
```

```python
from scipy import signal
sig = np.repeat([0., 1., 0.], 100)
win = signal.hann(50)
filtered = signal.convolve(sig, win, mode='same') / sum(win)

import matplotlib.pyplot as plt
fig, (ax_orig, ax_win, ax_filt) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original pulse')
ax_orig.margins(0, 0.1)
ax_win.plot(win)
ax_win.set_title('Filter impulse response')
ax_win.margins(0, 0.1)
ax_filt.plot(filtered)
ax_filt.set_title('Filtered signal')
ax_filt.margins(0, 0.1)
fig.tight_layout()
fig.show()
```
